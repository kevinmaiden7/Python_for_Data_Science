## Manejo de gráficas en Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

Utilizaremos la librería matplotlib para hacer algunos ejemplos de gráficas básicas:

$$ y = x $$

$$ y = x^2 $$

$$ y = x^3 $$

```python
x = np.linspace(-10,10,100)
x
```

```python
#Función y = ax + b
y = 0.5*x

plt.plot(x,y)
plt.ylim(0,5)
plt.xlim(0,10)
plt.ylabel('y = x', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.title(u"Gráfica de una función lineal\n",fontsize=16)
plt.show()
```

```python
y = x**2
plt.plot(x, y, color='orange')
plt.show()
```

```python
y = x**3
plt.plot(x, y)
plt.ylabel('$y = x^3$')
plt.xlabel('números entre -10 y 10')
plt.title(u'Ejemplos de función cúbica - UdeA\n')
plt.show()
```

Generando gráfica punteada

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y, 'r-')
plt.ylabel('y = seno(x)')
plt.xlabel('x')
plt.title('$s(t) = \mathcal{A}\sin(2 \omega t)$', fontsize=16, color='red')
plt.show()
```

Dos gráficas en el mismo plano

```python
y = np.sin(x)
y1 = np.cos(x)
plt.plot(y, 'r--', y1, 'k-')
plt.ylabel('y')
plt.xlabel('x')
plt.title(u'Ejemplos - UdeA\n')
plt.show()
```

Agregando el legend al gráfico

```python
y = np.sin(x)
y1 = np.cos(x)

plt.figure(figsize = (12,6)) # Change size
plt.plot(y, 'r-', label='Sin')
plt.plot(y1, 'k--', label='Cos')
plt.ylabel('y = seno(x)')
plt.xlabel('x')
plt.title(u'Ejemplos - UdeA\n')
plt.legend() # add it
# plt.legend([legend_1, legend_2])
plt.show()
```

Graficas de barras

```python
values = np.array([10, 50, 100])
labels = np.arange(3)

plt.bar(labels, values)
plt.title('Vertical Bar')
plt.show()
```

```python
values = np.array([10, 50, 100])
labels = np.arange(3)
plt.figure()

plt.barh(labels, values)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Horizontal Bar')
plt.show()
```

Gráficos para problemas de clasificación en Machine learning. Scatter plots

```python
#Creamos los datos artificiales

mu1, sigma1 = 1.5, 0.1
mu2, sigma2 = 1, 0.1
N = 100

x1 = mu1 + sigma1 * np.random.randn(N)
x1line = np.linspace(0,2,N)
x2 = mu2 + sigma2 * np.random.randn(N)
x2line = np.linspace(3,5,N)

#Decision boundary
b = -2
m = 3.5
v = np.linspace(0,100,100)
t = v*m + b

plt.scatter(x1, x1line, c='b')
plt.scatter(x2, x2line, c='r')
plt.plot(v, t, 'c-')
plt.ylim(-0.5, 6, 1)
plt.xlim(0.5, 1.8, 0.2)
plt.show()
```

También representar gráficas para visualizar relaciones matemáticas útiles

$P(x; \mu, \sigma) = \frac{1}{\sqrt{2\pi}\sigma }exp-\frac{1}{2}\frac{(x-\mu)^2}{\sigma^2}$





```python
#Generamos un conjunto de valores x
x = np.linspace(-3,3,100)
x
```

```python
mu, sigma = 0, 2

#los llevamos a la función de densidad de probabilidad normal
p = ( 1/(np.sqrt(2*np.pi)*sigma) ) * np.exp( (-1/2) * ( ((x - mu) ** 2) / (sigma ** 2) ) )

plt.plot(x, p, 'r--')
plt.show()
```
